```python
import matplotlib.pyplot as plt;
import numpy as np;

mydpi = 300;
numitems = 12000;

Design = np.array([20e3, 10e3, 100e3, 200e3]);
IC = np.array([10, 30, 2, 1]);
Assembly = np.array([10, 10, 2, 1]);
x = np.arange(1,numitems,1);

plt.figure(1);

Total = [];
for i in range(4):
    Total.append(Design[i] + (IC[i]+Assembly[i])*x);
    plt.plot(x,Total[i],linewidth=2,label=" "+str(i+1));

plt.grid();
plt.xlabel("");
plt.ylabel("  ()");
plt.xlim([0,numitems]);
plt.ylim([0,None]);
plt.legend();

plt.savefig("ex4totalpython.jpg",dpi=mydpi,bbox_inches="tight");
plt.show();



plt.figure(2);

PerUnit = [];
for i in range(4):
    PerUnit.append(Design[i]/x + (IC[i]+Assembly[i]));
    plt.plot(x,PerUnit[i],linewidth=2,label=" "+str(i+1));

plt.grid();
plt.xlabel("");
plt.ylabel("/ ()");
plt.xlim([0,numitems/4]);
plt.ylim([0,numitems/20]);
plt.legend();

plt.savefig("ex4perunitpython.jpg",dpi=mydpi,bbox_inches="tight");
plt.show();

```
